# Web Scraping Project

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_India"
```

```python
requests.get(url)
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text,'html')
```

```python
print(soup.prettify())
```

```python
soup.find_all('table')[0] ## find all returns list of all the tags required therefore we can use indexing
```

```python
soup.find('table',class_="wikitable sortable") ## normal way by specifying classname
```

```python
table = soup.find_all('table')[0]
```

```python
titles = table.find_all('th')
```

```python
table_titles = [title.text.strip() for title in titles]
```

```python
print(table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = table_titles)
```

```python
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    ind_row_data = [row_d.text.strip() for row_d in row_data]
    length = len(df)
    df.loc[length] = ind_row_data
```
